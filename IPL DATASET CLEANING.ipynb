```python
df = df.drop(columns=[
    'id',
    'season',
    'match_type',
    'super_over',
    'method',
    'umpire1',
    'umpire2'
    
])  #;removing irrelevant columns
```

```python
df['date'] = pd.to_datetime(df['date']) #changing date column datatype to date 
```

```python
df['city'] = df['city'].fillna(df['venue']) #handling missing values of city column
```

```python
df['target_runs'] = df['target_runs'].fillna(0)
df['target_overs'] = df['target_overs'].fillna(0)  #handling missing values of target runs and over column
```

```python
df['player_of_match'] = df['player_of_match'].fillna("Unknown") #handling missing values of player of the match column
```

```python
df['result_margin'] = df['result_margin'].fillna(0)  #handling missing values of result margin column
```

```python
df = df.dropna(subset=['winner']) #removing missing values of winner column
```

```python
df.isna().sum()
```

```python
team_map = {
    'Kings XI Punjab': 'Punjab Kings',
    'Royal Challengers Bangalore': 'Royal Challengers Bengaluru',
    'Rising Pune Supergiants' : 'Rising Pune Supergiant'
}

for col in ['team1', 'team2', 'winner', 'toss_winner']:
    df[col] = df[col].replace(team_map)
```

```python
df['team1'].unique()
```

```python
df['target_runs'] = df['target_runs'].astype(int)
df['target_overs'] = df['target_overs'].astype(int)
df['result_margin'] = df['result_margin'].astype(int) #changing datatype of this columns to integer
```
